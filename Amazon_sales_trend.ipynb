#Amazon Sales Trend

**Importing the Required Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Reading the given Dataset**

```python
table=pd.read_csv('Amazon Sales data.csv')
```

```python
table.head()
```

```python
table.shape
```

**Checking for any null values in the given dataset**

```python
table.isnull().sum()
```

**General information got from the given data**

```python
print("Total Revenue generated from 2010 to 2017 :",table['Total Revenue'].sum())
```

```python
print("Total Profit generated from 2010 to 2017 :",table["Total Profit"].sum())
```

```python
print("Total Cost from 2010 to 2017 :",table["Total Cost"].sum())
```

```python
print("Average Unit Cost :",table['Unit Cost'].mean())
```

```python
print("Average Unit Price :",table['Unit Price'].mean())
```

```python
print("Average Revenue generated per order :",table["Total Revenue"].mean())
```

```python
print("Average Profit got per order :",table["Total Profit"].mean())
```

```python
print("Average Cost per order :",table["Total Cost"].mean())
```

```python
print("Total Units Sold from 2010 to 2017 :",table['Units Sold'].sum())
```

```python
print("Average Units sold per order :",table['Units Sold'].mean())
```

Changing the "Order Date" column into datetime

```python
table['Order Date']=pd.to_datetime(table['Order Date'])
```

```python
table.columns
```

```python
table.head()
```

**Seperating month, year and month-year in a seperate columns**

```python
table['Order Month']=table['Order Date'].dt.month
```

```python
table['Order Year']=table['Order Date'].dt.year
```

```python
table['Order Year']=table['Order Year'].apply(str)
```

```python
table['Year-Month']=table['Order Date'].dt.to_period('M')
```

```python
table['Year-Month']=table['Year-Month'].apply(str)
```

```python
table.columns
```

```python
table=table[['Region', 'Country', 'Item Type', 'Sales Channel', 'Order Priority',
       'Order Date','Order Month', 'Order Year', 'Year-Month', 'Order ID', 'Ship Date', 'Units Sold', 'Unit Price',
       'Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit']]
```

```python
table.head()
```

***

**Monthly Sales Trend**

```python
monthly_sales=pd.DataFrame(table.groupby('Order Month')['Total Revenue'].sum())
```

```python
len(monthly_sales)
```

```python
monthly_sales['Month Name']=['January','February','March','April','May','June','July','August','September','October','November','December']
```

```python
monthly_sales
```

* Maximum Total revenue among all months from 2010 to 2017

```python
monthly_sales[monthly_sales['Total Revenue']==monthly_sales['Total Revenue'].max()]
```

* Minimum Total revenue among all months from 2010 to 2017

```python
monthly_sales[monthly_sales['Total Revenue']==monthly_sales['Total Revenue'].min()]
```

* Average Monthly sales

```python
avg_monthly_sales=monthly_sales['Total Revenue'].mean()
```

```python
print('Maximum monthly sales was in February :',monthly_sales['Total Revenue'].max())
```

```python
print('Minimum monthly sales was in August :',monthly_sales['Total Revenue'].min())
```

```python
print('Average Monthly Sales from 2010 to 2017 :',avg_monthly_sales)
```

```python
plt.figure(figsize=(13,6))
plt.title('Total Sales Across Months from 2010 to 2017',fontdict={"fontsize":15})
plt.bar(monthly_sales['Month Name'],monthly_sales['Total Revenue'])
plt.xlabel('Month',fontdict={"fontsize":12})
plt.ylabel('Total Revenue',fontdict={"fontsize":12})
plt.savefig('monthly.jpg')
plt.show()
```

* Now extracting and plotting the monthly profit from 2010 to 2017

```python
monthly_profit=pd.DataFrame(table.groupby('Order Month')['Total Profit'].sum())
```

```python
monthly_profit
```

```python
monthly_profit['Month Name']=['January','February','March','April','May','June','July','August','September','October','November','December']
```

```python
plt.figure(figsize=(13,6))
plt.title('Total Profit in all months from 2010 to 2017',fontdict={"fontsize":15})
plt.bar(monthly_profit['Month Name'],monthly_profit['Total Profit'])
plt.xlabel('Month',fontdict={"fontsize":13})
plt.ylabel('Total Profit',fontdict={"fontsize":13})
plt.savefig('monthly_profit.png')
plt.show()
```

* As February has the most monthly sales, lets examine febraury sales in all years

```python
feb_sales=table[table['Order Month']==2].reset_index()
```

```python
feb_yearly_sales=pd.DataFrame(feb_sales.groupby('Order Year')['Total Revenue'].sum())
```

```python
feb_yearly_sales
```

```python
plt.bar(feb_yearly_sales.index,feb_yearly_sales['Total Revenue'])
plt.title('Total Profit in all months from 2010 to 2017',fontdict={"fontsize":15})
plt.xlabel('Month',fontdict={"fontsize":13})
plt.ylabel('Total Profit',fontdict={"fontsize":13})
plt.savefig('monthly_profit.png')
plt.show()
```

**Now plotting bar graph of total revenue for all months (If any year was not mentioned in graph of a particular month then there were no sales in the month of that year)**

```python
plt.figure(figsize=(40,20))
for i in range(1,13):
  df_sales=table[table['Order Month']==i].reset_index()
  df=pd.DataFrame(df_sales.groupby('Order Year')['Total Revenue'].sum())
  plt.subplot(2,6,i)
  plt.title(monthly_sales['Month Name'][i],fontdict={"fontsize":15})
  plt.bar(df.index,df['Total Revenue'])
  plt.xlabel('Year',fontdict={"fontsize":12})
  plt.ylabel('Total Revenue',fontdict={"fontsize":12})
plt.savefig("all_month.png")
plt.show()
```
